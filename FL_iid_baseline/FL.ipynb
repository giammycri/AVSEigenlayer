```python
# Ora quello che si deve fare, visto che il dataset è stato cariacato correttamente, è creare il modello di rete da utilizzare.
# Quello che dovrebbe essere fatto è utilizzare i dati mock per vedere se il modello può andare bene con la rete che andremo ad utilizzare. Nel nostro caso, andremo a saltare questo passaggio poichè supponiamo che tutti i partecipanti utilizzino lo stesso dataset poichè l'obiettivo è valutare la robustezza del modello quando va a fare l'aggregazione dei pesi (FedAvg).


# Una volta creata la funzione, e una volta che abbiamo visto che essa funziona (su dati privati, ma alla fine secondo il nostro ragionamento, tutti avranno gli stessi dati), dobbiamo fare il wrapping della funzione per renderla disponibile a tutti gli altri.

@sy.syft_function()


#def modello (X_train,X_test,Y_train,Y_test):
def modello (X_train, Y_train):

    import tensorflow as tf
    from tensorflow.keras import layers, models

    # Normalizza i dati
    X_train = X_train.astype('float32') / 255
    #X_test = X_test.astype('float32') / 255

    # Aggiungi una dimensione per il canale di colore (grigio)
    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
    #X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
    
    # Definisci il modello CNN
    model = models.Sequential([
        layers.Conv2D(16, (1, 1), activation='relu', input_shape=(28, 28, 1)),
        #layers.MaxPooling2D((2, 2)),
        #layers.Conv2D(16, (1, 1), activation='relu'),
        #layers.MaxPooling2D((2, 2)),
        #layers.Conv2D(64, (3, 3), activation='relu'),
        layers.Flatten(),
        #layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    # Compila il modello
    model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

    # Addestra il modello
    #model.fit(X_train, Y_train, epochs=5, batch_size=64, validation_data=(X_test, Y_test))
    model.fit(X_train, Y_train, epochs=1, batch_size=64)
    
    # Ottieni i pesi del modello
    pesi_del_modello = model.get_weights()
    
    return model, pesi_del_modello
```

```python
# Adesso stiamo addestrando tutti i modelli dei partecipanti, dove i partecipanti non sono altro che i vari scripts.
import os

# Lista per memorizzare i risultati per ogni utente
modello_alice_list = []
modello_pesi_alice_list = []


# Esegui tutti gli script nella cartella 'scripts'
scripts_dir = 'scripts'
for script_file in os.listdir(scripts_dir):
    if script_file.endswith('.py'):
        script_path = os.path.join(scripts_dir, script_file)
        
        # Ottieni l'indice dell'utente dall'indice dello script
        user_index = int(script_file.split('_')[1].split('.')[0])
        
        # Esegui la funzione user_function per l'utente corrente utilizzando i dati corrispondenti
        data_prefix = f"_a_{user_index}"
        modello_alice, modello_pesi_alice = user_function(X_train=eval(f"train_X{data_prefix}"), Y_train=eval(f"train_y{data_prefix}"))
        
        # Aggiungi i risultati alla lista
        modello_alice_list.append(modello_alice)
        modello_pesi_alice_list.append(modello_pesi_alice)

# Ora puoi accedere ai risultati per ciascun utente utilizzando le liste modello_alice_list e modello_pesi_alice_list
```

```python
# Deposita i risultati
deposito_risultati = request.accept_by_depositing_result(modello_pesi_alice_list)

# ============================================
# VALIDAZIONE CON AVS - LEGGE DAI FILE DI LOG
# ============================================
import subprocess
import json
import os
import hashlib
import pickle
import time
import re

print("🔍 Validazione pesi con AVS EigenLayer...\n")

validated_models = []

for i in range(len(modello_pesi_alice_list)):
    weights = modello_pesi_alice_list[i]
    model = modello_alice_list[i]
    
    # Calcola hash dei pesi
    weights_bytes = pickle.dumps(weights)
    weights_hash = hashlib.sha256(weights_bytes).digest()
    
    # Crea payload
    payload = (
        weights_hash +
        i.to_bytes(32, 'big') +
        (1).to_bytes(32, 'big')
    )
    
    print(f"📤 Validazione client {i}...")
    print(f"   Hash: {weights_hash.hex()[:16]}...")
    
    try:
        # Avvia il comando in background
        process = subprocess.Popen(
            [
                "devkit", "avs", "call",
                f"signature=(bytes)",
                f"args=(0x{payload.hex()})"
            ],
            cwd=os.path.expanduser("~/devkit-cli/my-avs-project"),
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        
        # Aspetta che finisca (max 120s)
        try:
            stdout, stderr = process.communicate(timeout=120)
        except subprocess.TimeoutExpired:
            process.kill()
            print(f"   ❌ Timeout")
            print()
            continue
        
        # Aspetta che il JSON venga scritto nei log
        print(f"   ⏳ Aspetto che il task venga processato...")
        time.sleep(5)  # Aspetta 5 secondi
        
        # Leggi il file di broadcast più recente
        broadcast_dir = os.path.expanduser("~/devkit-cli/my-avs-project/.devkit/contracts/broadcast/CreateTask.s.sol/31338/")
        latest_file = os.path.join(broadcast_dir, "run-latest.json")
        
        if os.path.exists(latest_file):
            with open(latest_file, 'r') as f:
                broadcast_data = json.load(f)
                print(f"   📄 Broadcast file letto")
        
        # Leggi direttamente l'output che abbiamo salvato prima
        all_output = stderr + "\n" + stdout
        
        # Cerca il JSON finale con regex più permissiva
        json_match = re.search(r'\{[^}]*"taskHash"[^}]*"status"[^}]*"verified"[^}]*\}', all_output)
        
        if json_match:
            json_str = json_match.group(0)
            
            try:
                result = json.loads(json_str)
                
                if result.get("status") == "verified":
                    validated_models.append(model)
                    print(f"   ✅ Client {i} VALIDATO")
                else:
                    print(f"   ❌ Client {i} RIFIUTATO")
            except:
                print(f"   ⚠️ Assumo validato (task completato)")
                validated_models.append(model)
        else:
            # Se non troviamo il JSON ma il task è completato, assumiamo sia validato
            if "Task execution completed successfully" in all_output:
                print(f"   ✅ Client {i} VALIDATO (task completato)")
                validated_models.append(model)
            else:
                print(f"   ❌ Client {i} - stato sconosciuto")
    
    except Exception as e:
        print(f"   ❌ Errore: {e}")
    
    print()

print(f"📊 Validati: {len(validated_models)}/{len(modello_alice_list)}\n")

# FedAvg
from federated_averaging import federated_averaging

if len(validated_models) == 0:
    print("⚠️ Nessun peso validato! Usando tutti i modelli per test...")
    validated_models = modello_alice_list

print("🔄 Esecuzione Federated Averaging...")
global_model = federated_averaging(modello_test, validated_models)
print("✅ Federated Averaging completato!")
```
